```python
'''the below graphs are pdf(probability density function and cdf(cummulative distributive 
function))'''
counts,bin_edges=np.histogram(iris_setosa['PetalLengthCm'],bins=10)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'b--')
plt.plot(bin_edges[1:],cdf,'g')
counts1,bin_edges1=np.histogram(iris_virginca['PetalLengthCm'],bins=10)
pdf1=counts1/(sum(counts1))
print(pdf1)
print(bin_edges1)
cdf1=np.cumsum(pdf1)
plt.plot(bin_edges1[1:],pdf1,'b--')
plt.plot(bin_edges1[1:],cdf1,'g')
counts1,bin_edges1=np.histogram(iris_versicolor['PetalLengthCm'],bins=10)
pdf1=counts1/(sum(counts1))
print(pdf1)
print(bin_edges1)
cdf1=np.cumsum(pdf1)
plt.plot(bin_edges1[1:],pdf1,'b--')
plt.plot(bin_edges1[1:],cdf1,'g')
plt.show()
```
